```python
import pandas as pd
data_2022 = pd.read_csv("./data/데이터_2022.csv")
data_2023 = pd.read_csv("./data/데이터_2023.csv")
df_filtered = data_2022[data_2022['COAW_FLAG_NM'] != '참가증']
data_2023['COAW_FLAG_NM'].value_counts()

```
